### Statistical Analysis:
<br>
We also chose this question to perform some statistical analysis. We performed an ANOVA calculation and a Chi Square Test. 

11. Here we used the one-way ANOVA to perform predictive analysis on the independent variable represented by region, aand whether the dependent variable, the amount of bids, might influence how many cars are sold by each region. We broke up each region into its own group in order to perform the calculation. We are looking at whether there is a difference between the average amount of bids per region, in order to predict a difference in units sold. Our resulting p-score of 0.06 tells us there is no statistically significant difference between the means of our groups. 
<br>
<br>
The fact there is not a statistically significant difference bewteen the means of our groups is not evidence that a higher amount of bids will cause the sale of less cars, however knowing the amount of bids per region may not influence the amount of units sold can help us better predict region sales in the future. 
<br>
<br>
This result also makes sense, given what we know about the variable differences between the regions we are comparing - such as function, the amount of new dealers, and hotspots. We now have evidence that the number of bids alone, is not enough to predict the future numbers of units sold per region, because the ANOVA shows us the dependent variable is not significantly influencing the independent variable. 
<br>
<br>
12. We also performed the Chi Square Test, in order to compare the distribution of categorical data, i.e., bids per region.  Our results show a p-value of 0.0, indicating the significance is low. This falls in line with the previous results from the ANOVA test. We now have more evidence that the number of bids alone, is not enough to predict the future numbers of units sold per region.
